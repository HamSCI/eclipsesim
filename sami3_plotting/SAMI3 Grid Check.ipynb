```python
#!/usr/bin/env python3
import os
import cartopy.crs as ccrs

import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import scipy.io
```

```python
lat =   40.
lon = -100.

ml_fl_grid  = 'sami3/grid.mat'
ml_grid     = scipy.io.loadmat(ml_fl_grid)
ml_heights  = ml_grid['grid_heights']
ml_lats     = ml_grid['grid_lats']
ml_lons     = ml_grid['grid_lons']

lat_inx     = np.argmin(np.abs(ml_lats[:,0,0]-lat))
lon_inx     = np.argmin(np.abs(ml_lons[:,0,0]-lon))

alt_inx = 0
alt     = ml_heights[0,0,alt_inx]
```

```python
fig     = plt.figure(figsize=(15,6))
ax      = fig.add_subplot(111,projection=ccrs.PlateCarree())
ax.coastlines()
xx      = ml_lons[:,:,alt_inx]
yy      = ml_lats[:,:,alt_inx]
pcoll   = ax.pcolormesh(xx,yy,xx,vmin=-200,vmax=250)
#pcoll   = ax.scatter(xx,yy,c=xx,vmin=-180,vmax=180,s=100,edgecolor='face',marker='s')

cbar    = fig.colorbar(pcoll,label='Longitude')

xlim = (-150,-40)
ylim = (0,90)

ax.set_xlim(xlim)
ax.set_ylim(ylim)

txt     = '{:.1f} km Altitude'.format(alt)
fig.text(0.5,1,txt,fontdict={'weight':'bold','size':'x-large'},ha='center')
plt.show()
```
